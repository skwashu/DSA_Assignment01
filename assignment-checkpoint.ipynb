```python
# Delete the elements in a linked list whose sum is equal to zero

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def delete_zero_sum_sublists(head):
    dummy = Node(0)
    dummy.next = head
    prefix_sum = {0: dummy}
    curr_sum = 0
    
    while head:
        curr_sum += head.data
        if curr_sum in prefix_sum:
            prev = prefix_sum[curr_sum]
            prev.next = head.next
            while prev != head:
                curr_sum += prev.next.data
                del prefix_sum[curr_sum]
                prev.next = prev.next.next
            else:
                head = head.next
        else:
            prefix_sum[curr_sum] = head
            head = head.next
    return dummy.next


# Reverse a linked list in groups of given size

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        
def reverse_k_group(head, k):
    curr = head
    count = 0
    while curr and count < k:
        curr = curr.next
        count += 1
    
    if count == k:
        curr = reverse_k_group(curr, k)
        while count > 0:
            temp = head.next
            head.next = curr
            curr = head
            head = temp
            count -= 1
        head = curr
    
    return head


# Merge a linked list into another linked list at alternate positions

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def merge_alternate(head1, head2):
    curr1, curr2 = head1, head2
    while curr1 and curr2:
        next1, next2 = curr1.next, curr2.next
        curr1.next = curr2
        curr2.next = next1
        curr1, curr2 = next1, next2
    return head1

    
#Count pairs with a given sum in an array

def count_pairs_with_sum(arr, target_sum):
    count = 0
    freq = {}
    for num in arr:
        if num in freq:
            count += freq[num]
        complement = target_sum - num
        if complement in freq:
            freq[complement] += 1
        else:
            freq[complement] = 1
    return count


#Find duplicates in an array

def find_duplicates(arr):
    freq = {}
    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    return [num for num in freq if freq[num] > 1]


#Find the Kth largest and Kth smallest number in an array

def find_kth_smallest(arr, k):
    return sorted(arr)[k-1]

def find_kth_largest(arr, k):
    return sorted(arr, reverse=True)[k-1]


#Move all the negative elements to one side of the array

def move_negatives(arr):
    left, right = 0, len(arr) - 1
    while left <= right:
        if arr[left] < 0:
            left += 1
        else:
            arr[left], arr[right] = arr[right], arr[left]
            right -= 1
    return


#Reverse a string using a stack data structure

def reverse_string(s):
    stack = []
    for char in s:
        stack.append(char)
    reversed_s = ''
    while stack:
        reversed_s += stack.pop()
    return reversed_s


#Evaluate a postfix expression using stack

def evaluate_postfix(expression):
    stack = []
    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()
            if char == '+':
                stack.append(operand1 + operand2)
            elif char == '-':
                stack.append(operand1 - operand2)
            elif char == '*':
                stack.append(operand1 * operand2)
            elif char == '/':
                stack.append(operand1 // operand2) # Integer division for simplicity
    return stack.pop()

#Implement a queue using the stack data structure

class MyQueue:
    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x):
        self.in_stack.append(x)

    def pop(self):
        self.move_elements()
        return self.out_stack.pop()

    def peek(self):
        self.move_elements()
        return self.out_stack[-1]

    def empty(self):
        return not self.in_stack and not self.out_stack

    def move_elements(self):
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
```
